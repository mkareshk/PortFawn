```python
from portfawn.market_data import MarketData

%matplotlib inline
```

```python
## Parameters
# tickers
tickers = {"Global REIT": "REET", 
           "MSCI Emerging Markets": "EEM", 
           "Core S&P Total U.S. Stock Market": "ITOT",
           'Core U.S. Aggregate Bond': 'AGG'}

# date
start_date = "2015-01-01"
end_date   = "2020-12-30"

# misc
col_price = 'Close'
```

```python
## Create the market instance
market_data = MarketData(
    tickers = tickers,
    date_start = start_date,
    date_end = end_date,
    col_price = col_price,
)
```

## Price Data

```python
## Print the Price Data
market_data.data_prices.head(3)
```

```python
## Plot the Price Data
fig, ax = market_data.plot_prices()
```

## Returns Data

```python
## Print the Returns Data
market_data.data_returns.head(3)
```

```python
## Plot the Returns Data
fig, ax = market_data.plot_returns()
```

## cumulative Returns Data

```python
## Print the Returns Data
market_data.data_cum_returns.head(3)
```

```python
## Plot the Cum Returns Data
fig, ax = market_data.plot_cum_returns()
```

## Distributions of Returns

```python
# Statistics
market_data.data_returns.describe()
```

```python
## Dist of Returns
fig, ax = market_data.plot_dist_returns()
```

## Relationships of Returns

```python
# Cov Data
market_data.data_returns.cov()
```

```python
# Cov Plot
fig, ax = market_data.plot_cov()
```

```python
# Corr Data
market_data.data_returns.corr()
```

```python
# Corr
fig, ax = market_data.plot_corr()
```

## Expected Returns vs. Volatility

```python
# Expected Retuens - STD Data
market_data.mean_std
```

```python
# Expected Retuens - STD Plot
fig, ax = market_data.plot_mean_std(annualized=True)
```
