```python
from portfawn.market_data import MarketData

%matplotlib inline
```

```python
## Parameters
# tickers
tickers = {"DOW": "^DJI", 
              "NASDAQ": "^IXIC", 
              "RUSSELL2000": "^RUT", 
              "S&P500": "^GSPC", 
              "DAX": "^GDAXI", 
              "NIKKEI225": "^N225", 
              "HANG SENG": "^HSI", 
              "13 Week Treasury": "^IRX", 
              "30 Year Treasury": "^TYX"}
asset_list = list(tickers.values())

# date
start_date = "2021-01-01"
end_date = "2021-06-30"
```

```python
## Create the market instance
market_data = MarketData(
    tickers = tickers,
    date_start = start_date,
    date_end = end_date,
    col_price = "Close",
)
```

```python
## Print the Price Data
market_data.data_prices.head(10)
```

```python
## Plot the Price Data
fig = market_data.plot_prices()
```

```python
## Print the Returns Data
market_data.data_returns.head(10)
```

```python
## Plot the Returns Data
fig = market_data.plot_returns()
```

```python
## Plot the Cum Returns Data
fig = market_data.plot_cum_returns()
```

```python
## Dist of Returns
fig = market_data.plot_dist_returns()
```
